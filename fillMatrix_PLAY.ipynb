```python
"""10x10 matrix fill game play by hand"""

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a = np.full((10, 10), 0, dtype='int8')
a[0, 0] = 1
plt.imshow(a)
plt.show()

moves = 'edcxzaqw'
outcomes = ((-2, 2), (0, 3), (2, 2), (3, 0), (2, -2), (0, -3), (-2, -2), (-3, 0))
results = {move: outcomes[i] for i, move in enumerate(moves)}

i, j, k = 0, 0, 1
move_list = []

while True:
    
    valid = False
    
    while not valid:
        ii, jj = i, j
        m = input(f'next: {k}, move: ')

        if m not in moves:
            print(f'{m} not valid')
            continue

        ii += results[m][0]
        jj += results[m][1]
        if 0 <= ii < 10 and 0 <= jj < 10 and a[ii, jj] == 0:
            valid = True
        else:
            print('non valid')

    i = ii
    j = jj
    move_list.append(m)
    k = len(move_list)
    a[i, j] = k*2  # for vis
    plt.imshow(a)
    #plt.plot(j,i, '-r*',linewidth=2,)
    plt.show()

    a[i, j] = k
    print(k)

print(k, move_list)
```
